# What can I do with NLTK?

A brief introduction to what you can (and can't) do with NLTK

```python
import nltk
```

```python
from IPython.display import Image
Image(filename='tasks.png')
```

Things that are pretty difficult to do in natural language processing:
1. Understand complex emotions/sarcasm
2. Go from written text to a series of facts 
3. Machine Translation
4. Basically, understand text the way humans can understand text. 

I would group the main things NLTK can do into four categories:
1. Tokenising and counting those tokens
2. Part of speech tagging
3. Chunking and information extraction
4. Machine learning with text 

### Read in data

```python
f = open('reviews.txt', encoding="utf8")
raw = f.read()
```

```python
raw[0:500]
```

# 1. Tokenising 

### Word tokenising

```python
tokens = nltk.word_tokenize(raw)
```

```python
tokens[0:10]
```

### Sentence tokenising

```python
sentences = nltk.sent_tokenize(raw)
```

```python
sentences[0:5]
```

### Create text object

```python
text = nltk.Text(tokens)
```

```python
type(tokens)
```

```python
type(text)
```

### Finding the 10 most common "tokens".

```python
fdist = nltk.FreqDist(text)
```

```python
fdist.most_common(10)
```

### Finding common two word phrases.

```python
text.collocations()
```

 #### Regular Expressions
this game is so...

```python
text.findall(r'<this> <game> <is> <so> (<.*>)')
```

# 2. Part of speech tagging
See [Penn Treebank part of speech tags](www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html) to understand what each tag means. 

```python
tagged = nltk.pos_tag(text[0:1000])
```

```python
tagged[0:10]
```

### Finding the most common parts of speech

```python
tag_fd = nltk.FreqDist(tag for (word, tag) in tagged)
```

```python
tag_fd.most_common()[1:10]
```

### Most common nouns

```python
noun_fd = nltk.FreqDist(word for (word, tag) in tagged if tag[0] == 'N')
```

```python
noun_fd.most_common()[1:10]
```
